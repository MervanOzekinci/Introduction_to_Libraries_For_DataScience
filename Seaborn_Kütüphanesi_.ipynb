```python
#Seaborn Kütüphanesi

"""
Veri Görsellestirmenin önemini Matplotlib kütüphanesini aciklarken dile getirmistik. Fakat buraya direkt gecis yapmis olanlara tekrardan
bir aciklayalim.Normal sekilde sunulan verileri anlamak her zaman zor gelir insana.Ne demek zor!ben bir bakista 
bir grafigin dagilimini hemen söyleyebiliyorum diyen bir arkadas bi sakin ol. Ben veri derken 5-10 tane veriden bahsetmiyorum.Düsünün
ki elimizde bir dataseti var ve onlarca kategori altinda binlerce degerin dagilimini hayal etmek hayli bir zor hatta belki de imkansizdir.
Iste bunun icin biz elimize gelen onlarca,yüzlerce hatta binlerce veriden anlam cikarmak icin önce bunu bir görsellestirmemiz lazim ki,
bu verilerin dagilimana uygun olacak sekilde aklimizda bir yol ve model belirsin. Bunun icindir ki "matplotlib" ya da bize sunulan bir diger
kütüphane olan "seaborn" kullanbiliriz. Fakat biz profesyonel ve daha az kod ile daha güzel grafikler cizip bunu görsellestirmek istersek 
iste burada seaborn'u kullanmak gerekir.
""" 

```

```python
#kütphanlerin yüklenmesi
import seaborn as sns #seabornu sns seklinde cagir.Artik sns dedigimizde "seaborn"anlasilacak.Genel kullanim bicimi -sns-seklindedir.
import pandas as pd#pandas kütüphanesi artik pd olarak cagrilacak. 
```

#Simdi uzun vakit harcayip internetten dataset aramayalim. Biz hazir olan datasetlerden kullanalim
* load_dataset() fonksiyonu ile seaborn kütpnaesi icndeki "tips" datasetini yükleyecegiz
* head() >default olarak ilk 5 satiri gösterir. fakat biz head(10) dersek ilk 10 satiri gösterir. 
*tail(): ben bastan degil bitis yani sondan verileri görmek istiyorum diyenlerde tail() kullanabilir. Isteyen sondaki 10 satiri görmek 
isterse tail(10) yazabilir.

```python
#simdi seaborn icinde bulunan "tips" yükleyelim.Bu Verisetinin hikayesini bilmek bize anlamlandirma da katki saglayacaktir. Hadi anlamaya calisalim
#Bu dataseti gelen müsterilerin zaman,gün,cinsiyet,sigara icip icmeme gibi özellikleri göz önünde bulundurarak hazirlanmis bir veri setidir!
data=sns.load_dataset("tips")#sns icindeki tips verisetini data'ya atadik. Bu elimizde olan dataset üzerinde oynamalar yapacagimiz 
#icin bunun bir kopyasini alalim ve onun üzerinden islemler yapalim.Ne dersiniz :)

df=data.copy()#data yukarida tanimlanmisti, simdi üzerinden oynamalar yapmak icin bunun aynisinin kopyasini df'e atadik
df.head()#bunun ilk 5 satirini görelim


```

```python
#hocam bi dakka ben son 10 satirini görmek istioyrum diyenler hemen kizmayin, bi sakin olun, simdi onu gösterelim
df.tail(10)#alin size sondan 10 satiri
```

```python
#Simdi bunu noktasal gösterelim ve bu sefer bahşiş(tip) ve toplam hesab(total_bill) iliskisini inceleyelim.
#Unutmayin x,y sütün isimlerinden biri olmak zorundadir.Siz baska sütün isimleri ile de deneme yapabilirsiniz. 
#Ben burda merak ettim, acaba hesap arttikca bahşişte artiyor mu diye.
sns.relplot(
    data=df,
    x="total_bill",y="tip"
)
```

```python
#Simdi benim merakim biraz daha artti ve diyorum ki acaba bu bahşiş verenlerden en cok sigara mi icenler yoksa sigara icmeyenler mi?
#neden cünkü bahşiş verenler ,misal sigara icenler agirliktaysa, hizmet ederken en cok onlara yalakalik yapip daha fazla bahşiş kazanacagim :)
sns.relplot(
    data=df,
    x="total_bill",y="tip",hue="smoker"
)
#hue="smoker" sigara icip icmeyenleri farkli renklendir. Simdi grafige baktigimda görüyorum ki en cok bahşiş verenler sigara icmeyenler cogunlukta
#haaa,o zaman en cok bunlara güzel hizmet edeyim ki daha cok bahşiş biraksinlar. Artik sigara icmeyenler benim gözümde "candir" :)
```

```python
#Ya peki biz bir restaurant olarak gece gündüz acigiz,acaba sabah saatlerinde mi yoksa aksam saatlerinde mi en cok bahşiş birakiliyor?
#Nedenini aciklayacagim
sns.relplot(
    data=df,
    x="total_bill",y="tip",hue="smoker",col="time"
)

"""
col="time" burada col yani columns sütün ismi olarak zamana göre secim yapsin dedim.Bak ne görüyorum her iki durumda da en cok bahşişi 
birakanlar sigara icmeyenler ve aksam bahşişin daha cok birakildigini gördüm.O zaman hemen patrona söylüyorum beni geceye alsin diye.
Neden, cünkü bahşişi daha cok öyle kazanabilirim.  :) 

"""
```

```python
#Normalde ben haftasonu calismam ama ya daha cok para haftasonunda veriliyorsa diye bir kurt düstü icime hemen onunda analizini yapalim :)

sns.relplot(
    data=df,
    x="total_bill",y="tip",hue="smoker",col="day"
)

#Bak sen!, cumartes(sat) ve pazar(sun) daha cok bahşiş birakiliyor hemen ne yapiyorum? Analizimee göre gidip patrona haftasonu calismak 
#istedigimi söylüyorum. 
```

```python
#peki bakalim bahsis en cok kadinlar mi yoksa erkekler mi veriyor diye?Ama bu sefer farkl bir görsel olan barplot ile yapalim
sns.barplot(x="tip",y="sex",data=df)
#bu sefer görüyorum ki kadinlar daha az bahsis veriyor olmadi,üzdü bizi kadinlar ama yine de kadinlara hizmette sinir tanimayacagiz
#kapitalist düsünsekte "kadinlar cicektir" diyip hizmete devam. 
```

```python
sns.lmplot(data=df, x="total_bill", y="tip", col="time", hue="smoker")
#farkli bir gösterim yine de kendiliginden aciklayici olmus grafik.Sigara icmeyenlerin dogrusal egimi daha yüksek,yani bahsis onlarda hocam :)
```

```python
sns.displot(data=df, x="total_bill", col="time", kde=True)#Kde=True histogram egrisini gösterir.
#distplot() methodu tek degiskenli bir dagilimin hizlica gözlemlenmesinin en kolay yoludur.
```

```python
sns.displot(data=df, x="total_bill", col="time", kde=False)#Bakin burada egriyi cizmiyor tek fark bu
```

```python
sns.distplot(df.total_bill,hist = False);#Ilk defa sutun ismini df.total_bill ile cagrilabilecegini burada böylece görmüs olduk.
# hist=False yaparak da sadece yogunluk grafigini gösteririz.
```

```python
sns.boxplot(df.tip)#Boxplot ile verisetinin güvenilir araligini aslinda görüntülüyoruz.Ileride 1.5IQR alt ve üst ceyrek siniflarini
#gördügümüzde birkez daha deginebilirim oraya ama simdilik sunu bilmeliyiz ki genellikle bize gelen bahsisler 2 ile 4 arasinda 
#degisen bir araliktir. Ha bazen 6,8,10 bahsis birakanlar da var ama bunlar aykiri degerler. Dedigimiz gibi arada birakiyorlar :)
```

```python
sns.boxplot(x="sex",y="tip",hue="day",data=df);
#bak simdi burada cinsiyetlere göre degisen bahsislerin güne bagli degisimlerini inceliyoruz.
#Cok mükemmel degil mi ya, tek satirda bu kadar özelligi görsellestirmek :) 
```

```python
sns.boxplot(x="day",y="tip",hue="sex",data=df);#Burayi da siz yorumlayin :)
```

```python
sns.violinplot(x = "day", y = "total_bill", hue = "smoker", data = df)#tercihen kullanilir
```

```python
#Yine okuyuculardan kendi kendine acaba scatterplot var mi diyenlere de tabi ki var diye cevap verelim
sns.scatterplot(x="total_bill",y="tip",hue="smoker", data=df)#en bastaki relsplot ile ayni
```

```python
sns.catplot(x="day", y="tip", hue="smoker",data=df,kind="bar")#böyle güzel bir yani da bunun bar seklinde günlere bagli olarak
#sigara icip icmeyenlerin bahsis verme oranlarini görüyoruz.
```

```python
#simdi size pairplot gösterimini yapacagim fakat bunu baska bir veriseti ile yapacagim ki daha anlasilir olsun diye
#yine seaborn kütüphanesi icinde hazir gelen bir veriseti olan "penguins" ile baslayalim
penguins=sns.load_dataset("penguins")#verisetini dahil ettik

#simdide verilerimizin ne olduguna bi bakalim
penguins.head()
```

```python

sns.pairplot(penguins)

#pairplot ile birbirleri ile olan baglantilarini tek bir komutla halledebiliyoruz.Degerlerin grafikte dagilimlarina
#bakinca nasil bir iliski oldugunu görmek mümkün.
```

```python
sns.pairplot(penguins,hue="species")#species sütununa göre renklendirelim
```

```python
sns.pairplot(penguins,hue="species",kind="kde")#kde olarak göresllestirelim
```

```python
sns.pairplot(penguins,hue="species",diag_kind="hist")#diagram olarak da hist gösterilsin
```

```python
sns.pairplot(penguins,kind="reg")#grafige dogru eklemek istersek sayet bu durumda kind="reg" demeliyiz ki bize bunun egimi hakkinda bilgi versin
```

```python
#Simdi en son gösterecegim metodlardan olan heatmap analatacagim
#tabi bunun icin yine seaborn kütphanesinden bir veriseti yüklemem lazim ,verisetimiz:"fligts"
#zamana göre ucus yapan yolcularin sayisi hakkinda bize bilgi veren bir verisetidir,adindan da anlasilacagi gibi

flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")#pivot ile ay,yil ve yolcu sayilarini ekledik
sns.heatmap(flights)
```

```python
#Misal heatmap sol tarfina baktigimizda ve buna binen yolcu sayisini söylemek icin tam net olmayacak sekilde ama 
#1949,Jan diyebiliriz ki 100-200 civari, en sagda gösterilen rengin yogunluguna göre yolcu sayisini söylemek tam olmasa da mümkün
#fakat ben net olsun hersey bilinsin isterim diyorsaniz onun icinde bir yol var, haydi simdi onu gösterelim.
sns.heatmap(flights,annot=True,fmt="d",linewidths=0.5)

#annot=True> sayilarin görünmesini saglar
#fmt="d" >bu yazilan sayilarin okunabilir olmasini saglar
#linewidths=0.5 >kutucuklar arasina verilen deger kadar bosluk birakir


#böyle yapinca verilerin ne kadar net ve okunakli oldugunu görmus oluyoruz
```
