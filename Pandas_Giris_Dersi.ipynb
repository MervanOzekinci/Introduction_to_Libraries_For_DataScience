#Birlestirme(JOIN) Islemler

#Ileri Toplulastirma Islemleri(Aggregate,filter,transform,apply)

````python
import pandas as pd
````

````python
df=pd.DataFrame({'gruplar': ['A','B','C','A','B','C'],
                 'degisken1': [10,23,33,22,11,99],
                 'degisken2': [100,253,333,262,111,969]},
                 columns=['gruplar','degisken1','degisken2']
              )
````

````python
df
````

````python
#aggregate
````

````python
df.groupby('gruplar').mean()
````

````python
df.groupby('gruplar').aggregate([min,np.median,max])
````

````python
df.groupby('gruplar').aggregate({'degisken1':'min','degisken2':'max'})
#degisken1ve2 istenilen min ve max seklinde döndürülüp yazildi
````

#Filter

````python
df=pd.DataFrame({'gruplar': ['A','B','C','A','B','C'],
                 'degisken1': [10,23,33,22,11,99],
                 'degisken2': [100,253,333,262,111,969]},
                 columns=['gruplar','degisken1','degisken2']
)
df
````

````python
#kendi ihtiyacimiza göre fonksiyon yazacagiz
def filter_func(x):
  return x['degisken1'].std()>9
````

````python
df.groupby('gruplar').filter(filter_func)
````

````python
df.describe()
````

#transform

````python
df=pd.DataFrame({'gruplar': ['A','B','C','A','B','C'],
                 'degisken1': [10,23,33,22,11,99],
                 'degisken2': [100,253,333,262,111,969]},
                 columns=['gruplar','degisken1','degisken2']
              )
df
````

````python
df_a=df.iloc[:,1:3]
````

````python
df.transform(lambda x:x-x.mean())#bu drumda gruplar string oldugu icin hata olabilir
````

````python
df_a.transform(lambda x:(x-x.mean())/x.std())
````

#Pivot Tablolar

````python
import seaborn as sns
import pandas as pd
titanic=sns.load_dataset('titanic')
titanic.head()
````

````python
titanic.shape
````

````python
titanic.groupby('sex')['survived'].mean()#cinsiyete göre sag kalma durmunu ele aldik
````

````python
titanic.groupby('sex')[['survived']].mean()#cinsiyete göre sag kalma durmunu ele aldik
````

````python
titanic.groupby(['sex','class'])[['survived']].aggregate('mean').unstack()
````

````python
#pivot ile table
````

````python
titanic.pivot_table('survived',index='sex',columns='class')
#üstteki kodun daha basit yazilisi
````

````python
titanic.age.head()
````

````python
age=pd.cut(titanic['age'],[10,18,90])#yas araligini 10,18,90 araligina göre böldük
age.head(10)
````

````python
titanic.pivot_table('survived',['sex',age],'class')
````
