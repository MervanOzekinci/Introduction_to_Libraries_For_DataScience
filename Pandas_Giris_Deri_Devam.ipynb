#Gruplandirmak(Groupby)

```python
maasSozlugu={'Departman': ['Yazilim','IK','Yazilim','Hukuk','IK','Hukuk'],
             'Calisan Ismi': ['Ahmet','Mehmet','Yusuf','Kamil','Ayse','Sevil'],
             'Maas' :[100,150,120,244,124,44]}
```

```python
maasDataFrame=pd.DataFrame(maasSozlugu)
maasDataFrame
```

```python
groupObjesi=maasDataFrame.groupby('Departman')
```

```python
groupObjesi.count()#hangi departmanda kac kisi calisiyor
```

```python
groupObjesi.mean()#ayni departmanlarin ortalama verisi
```

```python
groupObjesi.max()
```

```python
groupObjesi.min()
```

```python
groupObjesi.describe()
```

#Concat

```python
sozluk1={'Isim':['Ahmet','Mehmet','Zeynep','Atil'],
         'Spor':['Kosu','Yüzme','Kosu','Basketbol'],
         'Kalori':[101,200,140,120]}
```

```python
df1=pd.DataFrame(sozluk1,index=[1,2,3,4])
df1
```

```python
sozluk2={'Isim':['Osman','Ali','Can','Seyma'],
         'Spor':['Kosu','Yüzme','Kosu','Basketbol'],
         'Kalori':[101,200,140,120]}
```

```python
df2=pd.DataFrame(sozluk2,index=[5,6,7,8])
df2
```

```python
sozluk3={'Isim':['Alex','Eva','Jana','Jannie'],
         'Spor':['Kosu','Tenis','Badminton','Basketbol'],
         'Kalori':[101,200,205,130]}
```

```python
df3=pd.DataFrame(sozluk3,index=[10,11,12,13])
```

#Concatenation>Birlestirme

```python
concat_df=pd.concat([df1,df2,df3])
concat_df
```

#Merge Islemi

```python
mergeSozluk={'Isim':['Ahmet','mehmet','zeynep','atil'],
             'Spor':['Kosu','Yüzme','Kosu','Basketbol']}
```

```python
 mergedf=pd.DataFrame(mergeSozluk)
 mergedf
```

```python
mergeSozluk2={'Isim':['Ahmet','mehmet','zeynep','atil'],
             'Kalori':[110,222,333,555]}
```

```python
mergedf2=pd.DataFrame(mergeSozluk2)
mergedf2
```

```python
pd.merge(mergedf,mergedf2, on='Isim')#ayni özellikleri tasiyan dataframleri ortak 
#özelliklerde birlestiriyor
```
