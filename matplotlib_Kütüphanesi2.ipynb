```python
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y)
plt.show()
```

```python
plt.title('ilk grafigim!')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y);
```

#Linewidth,lejand ekleme


```python
plt.plot(x,y, label='X^2',color='red',linewidth=4,linestyle='--',marker='+')
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.title('Ilk Grafigim')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

#Ayni grafikte birden fazla cizim yapmak

```python
plt.plot(x,y, label='X^2',color='red',linewidth=4,linestyle='--',marker='+')
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.title('Ilk Grafigim')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
#2.cizgimizi ciziyioruz
x2=np.arange(0,5,0.5)
plt.plot(x2,x2*2,color='green',linewidth=2,marker='.',label='2*x')
plt.legend()
```

#Grafigi Bilgisayarimiza kaydetme


```python
plt.plot(x,y, label='X^2',color='red',linewidth=4,linestyle='--',marker='+')
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.title('Ilk Grafigim')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
#2.cizgimizi ciziyioruz
x2=np.arange(0,5,0.5)
plt.plot(x2,x2*2,color='green',linewidth=2,marker='.',label='2*x')
plt.legend()
plt.savefig('ilkgrafik.png',dpi=400)
```
